```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Initializing the population
N = 100

A = np.zeros(int(N/2))
B = np.ones(int(N/2))

pop  = np.concatenate([A,B])
np.random.shuffle(pop)


r = 1.1
u = 0.01

j = 0

normalized_r_A = r / (1 + r)
normalized_r_B = 1 / (1 + r)

freq_of_B = np.array([])

for i in range(500000):
    
    reproduce = np.random.randint(0,N)
    kill      = np.random.randint(0,N)
    
    if pop[reproduce] == 0:
    
        rand1 = np.random.random()
        if rand1 < normalized_r_A:
            
            rand2 = np.random.random()
            if rand2 < u:
                
                pop[kill] = 1
                if i > j:
                    freq_of_B = np.append(freq_of_B,  np.sum(pop))
                    j = j + 50
    
                
            else:
                
                pop[kill] = 0
                if i > j:
                    
                    freq_of_B = np.append(freq_of_B, np.sum(pop))
                    j = j + 50
                
        continue
            
    else:
        
        rand = np.random.random()
        if rand < normalized_r_B:
            pop[kill] = pop[reproduce]
            
            if i > j:
                freq_of_B = np.append(freq_of_B, np.sum(pop))
                j = j + 50
            
        continue
        
```

```python
print("No. of type 1 in population = " + str(np.sum(pop)))
```

```python
#Plot

x = np.arange(0, freq_of_B.size)
y = freq_of_B
z = N - freq_of_B
plt.xlabel('Generation (scaled by a factor of 50)')
plt.ylabel('Frequency')
plt.plot(x,z)
plt.plot(x,y)
plt.legend(['Population of A', 'Population of B'])
plt.title('N = 100')
#plt.savefig('Plot.png')
```

```python
print("Theoretical frequency of Type 1 at equilibrium = 0.11 (as it is independant of N), which matches with the value from the simulation.")
```
