```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Initializing the population

N = 500

A = np.zeros(int(N/2))
B = np.ones(int(N/2))

pop  = np.concatenate([A,B])
np.random.shuffle(pop)

freq_of_B = np.array([])
```

```python
while (0 in pop and 1 in pop):
    
    #Two random individuals chosen from the population and the value at the kill index is changed to that at the reproduce
    #index effectively killing an individual and filling its slot with the reproduced individual. 
    
    reproduce = np.random.randint(0,500)
    kill      = np.random.randint(0,500)
    pop[kill] = pop[reproduce]
    
    #freq_of_B keeps track of indivuduals of B throught each generation
        
    freq_of_B = np.append(freq_of_B, np.sum(pop))
        
        
```

```python
x = np.arange(0, freq_of_B.shape[0])
y = freq_of_B
z = 500 - freq_of_B
plt.xlabel('Generation')
plt.ylabel('Frequency')
plt.plot(x,z)
plt.plot(x,y)
plt.legend(['Population of A', 'Population of B'])
plt.savefig('Moran.png')
```
